```python

import nltk
import operator
from nltk import tokenize
from operator import itemgetter
import math
nltk.download('stopwords')
nltk.download('punkt')





import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
stop_words = set(stopwords.words('english'))

def dothings(docName,doc):
    print("ruunning")
    total_words = tokenize.word_tokenize(doc)
    total_word_length = len(total_words)

    total_words = doc.split()
    total_word_length = len(total_words)


    total_sentences = tokenize.sent_tokenize(doc)
    total_sent_len = len(total_sentences)


    tf_score = {}
    for each_word in total_words:
        each_word = each_word.replace('.','')
        each_word = each_word.replace('--','')
        each_word = each_word.replace('—','')
        each_word = each_word.replace('–','')
        each_word = each_word.replace('-','')
        if each_word not in stop_words:
            if each_word in tf_score:
                tf_score[each_word] += 1
            else:
                tf_score[each_word] = 1

    # Dividing by total_word_length for each dictionary element
    tf_score.update((x, y/int(total_word_length)) for x, y in tf_score.items())
    #print(tf_score)



    def check_sent(word, sentences): 
        final = [all([w in x for w in word]) for x in sentences] 
        sent_len = [sentences[i] for i in range(0, len(final)) if final[i]]
        return int(len(sent_len))


    idf_score = {}
    for each_word in total_words:
        each_word = each_word.replace('.','')
        each_word = each_word.replace('--','')
        each_word = each_word.replace('—','')
        each_word = each_word.replace('–','')
        each_word = each_word.replace('-','')
        if each_word not in stop_words:
            if each_word in idf_score:
                idf_score[each_word] = check_sent(each_word, total_sentences)
            else:
                idf_score[each_word] = 1

    idf_score.update((x, math.log(int(total_sent_len)/y)) for x, y in idf_score.items())


 

    tf_idf_score = {key: tf_score[key] * idf_score.get(key, 0) for key in tf_score.keys()}



    def get_top_n(dict_elem, n):
        result = dict(sorted(dict_elem.items(), key = itemgetter(1), reverse = True)[:n]) 
        return result

    #%%

    # 11. Get the top 5 words of significance
    print("\n\n")
    print(docName)
    print(get_top_n(tf_idf_score, 15))
    print("\n\n")
# file1 = open("A  Murder-Suicide", encoding="utf8")
# #file1Read=file1.read()
# dothings(file1.read().lower(),1)#file1Read.lower(),1)
# file2 = open("A Festival of Books", encoding="utf8")
# dothings(file2.read().lower(),2)
# file3 = open("Better To Be Unlucky", encoding="utf8")
# file3Read=file3.read()

# dothings(file3Read.lower(),3)
# file4 = open("Cloning Pets", encoding="utf8")
# shishca=dothings(file4.read().lower(),4)
# file5 = open("Crazy Housing Prices", encoding="utf8")
# dothings(file5.read().lower(),5)
# file6 = open("Food Fight Erupted in Prison", encoding="utf8")
# dothings(file6.read().lower(),6)
# file7 = open("Freeway Chase Ends at Newsstand", encoding="utf8")
# dothings(file7.read().lower(),7)
# file8 = open("Gasoline Prices Hit Record High", encoding="utf8")
# dothings(file8.read().lower(),8)
# file9 = open("Happy and Unhappy Renters", encoding="utf8")
# dothings(file9.read().lower(),9)
# file10 = open("Jerry Decided To Buy a Gun", encoding="utf8")
# dothings(file10.read().lower(),10)
# file12 = open("Man Injured at Fast Food Place", encoding="utf8")
# dothings(file12.read().lower(),11)
# file13 = open("Pulling Out Nine Tons of Trash", encoding="utf8")
# dothings(file13.read().lower(),12)
# file14 = open("Rentals at the Oceanside Community", encoding="utf8")
# dothings(file14.read().lower(),13)
# file15 = open("Sara Went Shopping", encoding="utf8")
# dothings(file15.read().lower(),14)
# file16 = open("Trees Are a Threat", encoding="utf8")
# dothings(file16.read().lower(),15)
# with open("A  Murder-Suicide",'r') as a:
#     print(a.read())

#most keywords have no correlation to the actual document given. some are like food fight erupted in prison was mentioned "inmate" but that is all. 
```
